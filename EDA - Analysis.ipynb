```python
#lihat tanggal

supermarket['Date'] = pd.to_datetime(supermarket['Date'])
supermarket['day'] = (supermarket['Date']).dt.day
supermarket['month'] = (supermarket['Date']).dt.month
supermarket['year'] = (supermarket['Date']).dt.year
```

```python
# lihat waktu

supermarket['Time'] = pd.to_datetime(supermarket['Time'])
supermarket['Hour'] = (supermarket['Time']).dt.hour  
```

```python
city = supermarket['City']
sns.countplot(city,order=city.value_counts().index)
```

Supermarket di 3 kota diatas customernya hampir seimbang, Yangon 34%, Mandalay 33,2%, dan kota Naypyitaw 32,8%

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
fig.set_size_inches(12,8)

yangon = supermarket[supermarket['City']=='Yangon']['Customer type']
mandalay = supermarket[supermarket['City']=='Mandalay']['Customer type']
naypyitaw = supermarket[supermarket['City']=='Naypyitaw']['Customer type']

sns.countplot(yangon,ax=ax1, order =yangon.value_counts().index )
ax1.set_title('Yangon')
sns.countplot(mandalay,ax=ax2,order =mandalay.value_counts().index )
ax2.set_title('Mandalay')
sns.countplot(naypyitaw,ax=ax3, order =naypyitaw.value_counts().index)
ax3.set_title('Naypyitaw')
```

Diantara 3 kota diatas customernya hampir sebanding antara yang bukan member dan member.

```python
fig, (ax1, ax2) = plt.subplots(2, 1,figsize=(24,6*3))
sns.countplot(data = supermarket[supermarket['Customer type']=='Member'],x='City',hue='Payment',ax=ax1)
ax1.set_title('Member - Payment')
sns.countplot(data = supermarket[supermarket['Customer type']=='Normal'],x='City',hue='Payment',ax=ax2)
ax2.set_title('Normal - Payment')
```

Member dikota Yangon lebih banyak membayar menggunakan metode pembayaran Ewallet, sedangkan
Member dikota Mandalay lebih banyak membayar menggunakan metode pembayaran Credit Card, dan
Member dikota Naypyitaw customer yang menggunakan metode pembayaran Credit card dan cash hampir sama.

Untuk Customer yang bukan member dikota Naypyitaw paling banyak menggunakan metode pembayaran cash,sedangkan
dikota Yangon dan dikota Mandalay customer paling banyak menggunakan metode pembayaran Ewallet

```python
ewallet = supermarket[(supermarket['City']=='Yangon') & (supermarket['Payment']=='Ewallet')]['Product line']
cash = supermarket[(supermarket['City']=='Yangon') & (supermarket['Payment']=='Cash')]['Product line']
cc = supermarket[(supermarket['City']=='Yangon') & (supermarket['Payment']=='Credit card')]['Product line']

fig, (ax1, ax2, ax3) = plt.subplots(3, 1)
fig.set_size_inches(17, 19)
sns.countplot(ewallet,ax=ax1,order=ewallet.value_counts().index)
ax1.set_title('Product line - Metode Pembayaran Ewallet')
sns.countplot(cash,ax=ax2,order=cash.value_counts().index)
ax2.set_title('Product line - Metode Pembayaran Cash')
sns.countplot(cc,ax=ax3,order=cc.value_counts().index)
ax3.set_title('Product line - Metode Pembayaran Credit Card')
```

Dikota Yangon customer yang belanja menggunakan Ewallet dan Cash paling banyak belanja produk Home and Lifestyle, sedangkan pembayaran dengan metode Credit Card paling banyak belanja produk Food and beverages dan electronic accessories.

```python
ewallet = supermarket[(supermarket['City']=='Mandalay') & (supermarket['Payment']=='Ewallet')]['Product line']
cash = supermarket[(supermarket['City']=='Mandalay') & (supermarket['Payment']=='Cash')]['Product line']
cc = supermarket[(supermarket['City']=='Mandalay') & (supermarket['Payment']=='Credit card')]['Product line']

fig, (ax1, ax2, ax3) = plt.subplots(3, 1)
fig.set_size_inches(17, 19)
sns.countplot(ewallet,ax=ax1,order=ewallet.value_counts().index)
ax1.set_title('Product line - Metode Pembayaran Ewallet')
sns.countplot(cash,ax=ax2,order=cash.value_counts().index)
ax2.set_title('Product line - Metode Pembayaran Cash')
sns.countplot(cc,ax=ax3,order=cc.value_counts().index)
ax3.set_title('Product line - Metode Pembayaran Credit Card')
```

Customer dikota Mandalay paling banyak belanja produk home and lifestyle menggunakan metode pembayaran Ewallet, untuk metode pembayaran Cash paling banyak membeli produk Sports and Travel dan untuk metode pembayaran credit card produk yang paling banyak dibeli adalah Food and beverages

```python
ewallet = supermarket[(supermarket['City']=='Naypyitaw') & (supermarket['Payment']=='Ewallet')]['Product line']
cash = supermarket[(supermarket['City']=='Naypyitaw') & (supermarket['Payment']=='Cash')]['Product line']
cc = supermarket[(supermarket['City']=='Naypyitaw') & (supermarket['Payment']=='Credit card')]['Product line']

fig, (ax1, ax2, ax3) = plt.subplots(3, 1)
fig.set_size_inches(17, 19)
sns.countplot(ewallet,ax=ax1,order=ewallet.value_counts().index)
ax1.set_title('Product line - Metode Pembayaran Ewallet')
sns.countplot(cash,ax=ax2,order=cash.value_counts().index)
ax2.set_title('Product line - Metode Pembayaran Cash')
sns.countplot(cc,ax=ax3,order=cc.value_counts().index)
ax3.set_title('Product line - Metode Pembayaran Credit Card')
```

Dikota Naypyitaw kebanyakan menggunakan metode pembayaran Ewallet dan credit card untuk membayar produk Fashion accessories , untuk metode pembayaran cash produk yang banyak dibeli adalah food and beverages
